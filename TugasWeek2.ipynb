```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img5 = cv.imread('/content/drive/MyDrive/Kuliah/PCVK/Images/kitten01.jpg')
img6 = cv.cvtColor(img5, cv.COLOR_BGR2RGB)
R,G,B = cv.split(img6)
zeros = np.zeros(img6.shape[:2], dtype = "uint8")

red_green = cv.merge([R,G,zeros])
plt.imshow(red_green)
```

```python
green_blue = cv.merge([zeros,G,B])
plt.imshow(green_blue)
```

```python
from random import randrange
img10 = cv.imread('/content/drive/MyDrive/Kuliah/PCVK/Images/kitten01.jpg')
img11 = cv.cvtColor(img10, cv.COLOR_BGR2RGB)
t = img11.shape[0]
l = img11.shape[1]

for x in range(5):
  cv.rectangle(img11, (randrange(l),randrange(t)), (randrange(l),randrange(t)), (randrange(255),randrange(255),randrange(255)), 5)
plt.imshow(img11)
```
